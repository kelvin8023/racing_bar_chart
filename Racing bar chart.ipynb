```python
import pandas as pd 
import bar_chart_race as bcr #package for creating the racin gbar chart
```

```python
data = pd.read_csv("Target.csv")
```

```python
data.head(10)
```

```python
print([a for a in data.columns])
```

```python
data_simplified = data[['create_date', 'destination_location_name','booked_ffes']]
data_simplified.set_index('create_date') #need to set the date(time series as the index)
```

```python
#data_simplified.isnull().values.any() #it can not contain any nan value in the data
```

```python
table = data_simplified.pivot_table(index=['create_date'], columns='destination_location_name', values='booked_ffes') # need to flatten the dataframe to stretch the identifier to the colomn
table
```

```python
flattened = pd.DataFrame(table.to_records()) # convert the pivot table to dataframe
type(flattened)
```

```python
#flattened.fillna(0)
flattened.set_index('create_date', inplace = True)
```

```python
cum_sum_df = flattened.cumsum(axis = 0)
cum_sum_df.tail(10)
cum_sum_df.fillna(0)
```

```python
bcr.bar_chart_race(df =cum_sum_df,filename=None,figsize = (4,5))
```
